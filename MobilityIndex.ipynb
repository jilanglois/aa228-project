```python
data = pandas.read_csv("data/2020_US_Region_Mobility_Report.csv")
data.set_index(["country_region", "sub_region_1", "sub_region_2", "date"], inplace=True)
var = data.columns[-6:]

data_scc = data.loc["United States"].loc["California"].loc["Santa Clara County"][var]

import plotly.express as px
from plotly import graph_objects as go

fig = go.Figure()
for i in range(len(var)):
    fig.add_trace(go.Scatter(x=data_scc.index, y=data_scc[var[i]].rolling(7).mean()/100., name=" ".join(var[i].split("_")[:-4])))
fig.update_yaxes(tickformat="%")
fig.show()
```

```python
import plotly.express as px
from plotly import graph_objects as go

fig = go.Figure()
for i in range(len(var)):
    fig.add_trace(go.Scatter(x=data_scc.index, y=data_scc[var[i]].rolling(7).mean()/100., name=" ".join(var[i].split("_")[:-4])))
fig.update_yaxes(tickformat="%")
fig.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(data_scc)

n_components=len(var)
pca = PCA(n_components=n_components)
pca_rows = ["pc%i" % (i+1) for i in range(n_components)]
x_ = pca.fit_transform(x)

pandas.DataFrame(pca.explained_variance_ratio_.reshape([1, len(pca_rows)]),
                                                       index=["ExplaiendVarianceRatio"],
                                                       columns = pca_rows)
```

```python
pandas.DataFrame(pca.components_, index=pca_rows, columns=var)
```

```python
mobility_index = pandas.DataFrame(x_[:,0], index=data_scc.index, columns=["mobility_index"]).rolling(7).mean()
```

```python
import plotly.express as px

fig = px.line(mobility_index.reset_index(), x="date", y="mobility_index")
fig.show()
```
